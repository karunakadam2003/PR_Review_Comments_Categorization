```python
import re
import matplotlib.pyplot as plt
from collections import Counter
import google.generativeai as genai
import pandas as pd
from tabulate import tabulate


genai.configure(api_key="API-Key Here")


comments = [
    "This function is not handling edge cases properly.",
    "There is a null pointer exception risk here.",
    "The error handling logic is missing in this function.",
    "Boundary conditions should be tested to avoid unexpected failures.",
    "This logic will fail when the input is negative.",
    "Consider using a more efficient algorithm here.",
    "The variable names are not meaningful, making the code hard to read.",
    "Try to optimize this loop to improve performance.",
    "Use consistent indentation and formatting for better readability.",
    "This function is too long. Consider breaking it into smaller methods.",
    "Avoid using magic numbers; use constants instead.",
    "There are redundant conditions in this if-else block.",
    "There's a potential security vulnerability in this function.",
    "This input is not properly sanitized, leading to a possible injection attack.",
    "Consider encrypting sensitive data before storing it.",
    "Use HTTPS instead of HTTP for secure communication.",
    "Ensure authentication checks are in place before performing this action.",
    "Please update the documentation for this method.",
    "Add a docstring to describe the purpose of this function.",
    "The parameter descriptions are missing in the method docstring.",
    "This comment is outdated and does not match the current implementation.",
    "Consider adding inline comments to explain complex logic.",
    "Try to optimize this loop to improve performance.",
    "This query can be optimized by adding an index.",
    "Avoid recalculating the same value inside a loop.",
    "Consider using vectorized operations instead of iterating over elements.",
    "This recursive function may cause a stack overflow for large inputs.",
    "There are no unit tests for this function. Please add them.",
    "This test case does not cover edge scenarios.",
    "Consider using mock data instead of real data in test cases.",
    "The assertions in this test are too weak. Make them more specific.",
    "Ensure proper teardown methods to clean up resources after tests.",
    "This code is difficult to maintain. Consider refactoring it.",
    "Avoid deep nesting, as it reduces code readability.",
    "Separate business logic from UI logic for better separation of concerns.",
    "Use dependency injection instead of hardcoded dependencies.",
    "Avoid committing commented-out code; remove unused code instead.",
]


categories = {
    "Bug": ["error", "bug", "issue", "edge case", "fix", "exception", "crash", "failure"],
    "Code Quality": ["readability", "refactor", "naming", "clean code", "indentation", "redundant", "duplicate"],
    "Security": ["vulnerability", "secure", "risk", "attack", "encryption", "injection", "authentication", "XSS"],
    "Documentation": ["comment", "docs", "documentation", "explanation", "docstring", "inline comment", "param"],
    "Performance": ["optimize", "slow", "efficient", "performance", "speed", "query", "vectorized", "cache"],
    "Testing": ["unit test", "mock", "assertion", "test case", "coverage", "edge scenarios", "teardown"],
    "Maintainability": ["maintainability", "refactor", "nesting", "dependency injection", "hardcoded", "separation"],
}

def categorize_comment(comment):
    """Categorize the comment using Gemini AI or fallback to keyword matching."""
    try:
        MODEL_NAME = "gemini-2.0-flash-lite"
        model = genai.GenerativeModel(MODEL_NAME)
        response = model.generate_content(f"Categorize this PR comment: '{comment}' into Bug, Code Quality, Security, Documentation, or Performance.")
        print(f"Categorized using {MODEL_NAME}")
        category = response.text.strip()
        if category in categories:
            return category
    except Exception as e:
        print(f"AI categorization failed: {e}")

    # Fallback to keyword-based classification
    for category, keywords in categories.items():
        if any(re.search(rf"\b{word}\b", comment, re.IGNORECASE) for word in keywords):
            return category
    return "Other"

categorized_data = [(comment, categorize_comment(comment)) for comment in comments]
categorized_comments = [category for _, category in categorized_data]
category_counts = Counter(categorized_comments)

# Calculate total PR comments for percentage calculation
total_comments = len(comments)
category_percentage = {category: (count / total_comments) * 100 for category, count in category_counts.items()}

# Convert to DataFrame for better visualization
df = pd.DataFrame(categorized_data, columns=["PR Review Comment", "Category"])

# Display table with count and percentage
summary_data = [(category, count, f"{percentage:.2f}%") for category, count, percentage in zip(category_counts.keys(), category_counts.values(), category_percentage.values())]
summary_df = pd.DataFrame(summary_data, columns=["Category", "Count", "Percentage"])

print("\nðŸ”¹ PR Review Categorization Summary:")
print(tabulate(summary_df, headers="keys", tablefmt="pretty"))

# Plot analytics
fig, axs = plt.subplots(1, 2, figsize=(14, 6), facecolor='#f5f5f5')

# Bar chart with count and percentage annotations
colors = ['#3498db', '#2ecc71', '#e74c3c', '#9b59b6', '#f39c12', '#1abc9c', '#e67e22']
bars = axs[0].bar(category_counts.keys(), category_counts.values(), color=colors, edgecolor='black', linewidth=1.2)

axs[0].set_title("PR Review Comment Categories", fontsize=14, fontweight='bold')
axs[0].set_ylabel("Count", fontsize=12)
axs[0].set_xlabel("Category", fontsize=12)
axs[0].tick_params(axis='x', rotation=30)
axs[0].grid(axis='y', linestyle='--', alpha=0.7)

# Annotate bars with percentages
for bar, (category, percentage) in zip(bars, category_percentage.items()):
    height = bar.get_height()
    axs[0].text(bar.get_x() + bar.get_width() / 2, height, f"{percentage:.1f}%", ha='center', va='bottom', fontsize=10, fontweight='bold')

# Pie chart with count and percentage labels
axs[1].pie(category_counts.values(), labels=[f"{cat} ({pct:.1f}%)" for cat, pct in category_percentage.items()],
           autopct='%1.1f%%', colors=colors, wedgeprops={'edgecolor': 'black', 'linewidth': 1.2}, textprops={'fontsize': 10})
axs[1].set_title("Category Distribution", fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import re
import matplotlib.pyplot as plt
from collections import Counter
import google.generativeai as genai
import pandas as pd
from tabulate import tabulate

# Configure Google Gemini API
genai.configure(api_key="API-Key here")

# Sample PR review comments
comments = [
    "This function is not handling edge cases properly.",
    "There is a null pointer exception risk here.",
    "The error handling logic is missing in this function.",
    "Consider using a more efficient algorithm here.",
    "There is a potential security vulnerability in this function.",
    "Please update the documentation for this method.",
    "Try to optimize this loop to improve performance.",
    "There are no unit tests for this function. Please add them.",
    "This code is difficult to maintain. Consider refactoring it.",
]

# Define categories
categories = {
    "Bug": ["error", "bug", "issue", "edge case", "fix", "exception", "crash", "failure"],
    "Code Quality": ["readability", "refactor", "naming", "clean code", "indentation", "redundant", "duplicate"],
    "Security": ["vulnerability", "secure", "risk", "attack", "encryption", "injection", "authentication"],
    "Documentation": ["comment", "docs", "documentation", "explanation", "docstring", "inline comment", "param"],
    "Performance": ["optimize", "slow", "efficient", "performance", "speed", "query", "vectorized", "cache"],
    "Testing": ["unit test", "mock", "assertion", "test case", "coverage", "edge scenarios", "teardown"],
    "Maintainability": ["maintainability", "refactor", "nesting", "dependency injection", "hardcoded", "separation"],
}

def batch_categorize_comments(comments):
    """Categorize multiple PR comments using a single Gemini API request."""
    try:
        MODEL_NAME = "gemini-2.0-flash-lite"
        model = genai.GenerativeModel(MODEL_NAME)

        prompt = f"""
        Categorize the following PR review comments into one of these categories: Bug, Code Quality, Security, Documentation, Performance, Testing, Maintainability.

        Return the response in JSON format with "comment" and "category" keys.

        PR Comments:
        {comments}

        Example Response:
        [
            {{"comment": "This function is not handling edge cases properly.", "category": "Bug"}},
            {{"comment": "Consider using a more efficient algorithm here.", "category": "Code Quality"}}
        ]
        """

        response = model.generate_content(prompt)

        # Extract structured JSON output
        import json
        categorized_data = json.loads(response.text)

        # Convert response into a dictionary for quick lookup
        categorized_dict = {item["comment"]: item["category"] for item in categorized_data}
        return [categorized_dict.get(comment, "Other") for comment in comments]

    except Exception as e:
        print(f"AI categorization failed: {e}")
        return ["Other"] * len(comments)

# Categorize comments in batch
categorized_comments = batch_categorize_comments(comments)

# Convert to DataFrame
df = pd.DataFrame({"PR Review Comment": comments, "Category": categorized_comments})

# Count category occurrences
category_counts = Counter(categorized_comments)

# Calculate percentages
total_comments = sum(category_counts.values())
category_percentages = {category: f"{(count/total_comments) * 100:.1f}%" for category, count in category_counts.items()}

# Print categorized data
print(tabulate(df, headers='keys', tablefmt='pretty'))

# Print category distribution with percentages
print("\nCategory Distribution:")
for category, count in category_counts.items():
    print(f"{category}: {count} PRs ({category_percentages[category]})")

# Plot analytics
fig, axs = plt.subplots(1, 2, figsize=(14, 6), facecolor='#f5f5f5')

# Bar chart with count and percentage
axs[0].bar(category_counts.keys(), category_counts.values(), color=['#3498db', '#2ecc71', '#e74c3c', '#9b59b6', '#f39c12'], edgecolor='black', linewidth=1.2)
axs[0].set_title("PR Review Comment Categories", fontsize=14, fontweight='bold')
axs[0].set_ylabel("Count", fontsize=12)
axs[0].set_xlabel("Category", fontsize=12)
axs[0].tick_params(axis='x', rotation=30)
axs[0].grid(axis='y', linestyle='--', alpha=0.7)

# Pie chart with percentage labels
axs[1].pie(category_counts.values(), labels=[f"{cat} ({category_percentages[cat]})" for cat in category_counts.keys()], autopct='%1.1f%%',
           colors=['#3498db', '#2ecc71', '#e74c3c', '#9b59b6', '#f39c12'],
           wedgeprops={'edgecolor': 'black', 'linewidth': 1.2}, textprops={'fontsize': 10})
axs[1].set_title("Category Distribution", fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```
